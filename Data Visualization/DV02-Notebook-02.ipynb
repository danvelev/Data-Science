## Practice
Recreating the example given in the assignment in order to get familliar with the Matplotlib functions

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

#to show plots inline with the iPython Notebook
%matplotlib inline

X = np.linspace(-np.pi, np.pi, 100) # Array 100 point in the range of -pi until pi
Y = np.sin(X)

plt.title('First plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(-4,4)

plt.plot(X,Y)

plt.show();
```

## Assignment
Create a plot with a given lines, which contains multiple subplots.

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

%matplotlib inline

fig, ax = plt.subplots()
ax.plot(X, Y, 'r--', linewidth = 2)
ax.plot(X, Y/2, 'b-', linewidth = 2)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Subplot')
plt.ylim(-2.0, 2.0) # to test whether the lines zoom out or not
plt.legend(['red curve', 'blue curve'], loc = 'best')

plt.show()
```
