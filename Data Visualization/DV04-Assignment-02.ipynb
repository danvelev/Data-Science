## Solution

"In July, 1995, a scorching heat wave hit Chicago, killing seven hundred and thirty-nine people, roughly seven times as many as died in Superstorm Sandy. Soon after the heat abated, social scientists began to look for patterns behind the deaths. Some of the results were unsurprising: having a working air-conditioner reduced the risk of death by eighty per cent. But fascinating patterns did emerge.For the most part, the geography of heat-wave mortality was consistent with the cityâ€™s geography of segregation and inequality: eight of the ten community areas with the highest death rates were virtually all African-American, with pockets of concentrated poverty and violent crime, places where old people were at risk of hunkering down at home and dying alone during the heat wave. At the same time, three of the ten neighborhoods with the lowest heat-wave death rates were also poor, violent, and predominantly African-American. "

We have decided to make a comparison between the number of deaths( grouped by gender, race, age) measured after the two big disasters, Heat Wave in Chicago and Superstorm Sandy. Although that Sandy made much more damages to the properties, it has less victims. We take under consideration that during the heat wave, people without air conditions in their homes are more likely to suffer.

As we can see in the graphs below, in deaths by race, there were reported more dead black people than white, due to the poverty of neighbourhoods. Another fact that we can conclude is that the people in the age group between 70 and 80 could not handle the heat.

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline

# loading data
df = pd.read_csv('data/data-by-gender-heat.csv')
df_sandy = pd.read_csv('data/data-by-gender-sandy.csv')
# df_disasters = pd.concat([df,df_sandy], axis=1, ignore_index=True)

# creating the barplot
sns.set_style("whitegrid")
ax = sns.barplot(x="Gender", y="Number of Deaths", data=df, color="salmon")
ax.set_title("Deaths by Gender")
ax1 = sns.barplot(x="Gender", y="Number of Deaths", data=df_sandy, color="sage")

# creating the legend, using pyplot
topbar = plt.Rectangle((0,0),1,1, fc="salmon", edgecolor='none')
bottombar = plt.Rectangle((0,0),1,1, fc="sage", edgecolor='none')
l = plt.legend([topbar, bottombar], ['Heat Wave Chicago', 'Sandy'], loc=1, ncol=2, prop={'size':10})
l.draw_frame(False)
```

```python
# loading data
df = pd.read_csv('data/data-by-race-heat.csv')
df_sandy = pd.read_csv('data/data-by-race-sandy.csv')
# df_disasters = pd.concat([df,df_sandy], axis=1, ignore_index=True)

# creating the barplot
sns.set_style("whitegrid")
ax = sns.barplot(x="Race", y="Number of Deaths", data=df, color="salmon")
ax.set_title("Deaths by Race")
ax1 = sns.barplot(x="Race", y="Number of Deaths", data=df_sandy, color="sage")

# creating the legend, using pyplot
topbar = plt.Rectangle((0,0),1,1, fc="salmon", edgecolor='none')
bottombar = plt.Rectangle((0,0),1,1, fc="sage", edgecolor='none')
l = plt.legend([topbar, bottombar], ['Heat Wave Chicago', 'Sandy'], loc=1, ncol=2, prop={'size':10})
l.draw_frame(False)
```

```python
# loading data
df = pd.read_csv('data/data-by-age-heat.csv')
df_sandy = pd.read_csv('data/data-by-age-sandy.csv')
# df_disasters = pd.concat([df,df_sandy], axis=1, ignore_index=True)

# creating the barplot
sns.set_style("whitegrid")
ax = sns.barplot(x="Age Group", y="Number of Deaths", data=df, color="sage")
ax.set_title("Deaths by Age")
ax1 = sns.barplot(x="Age Group", y="Number of Deaths", data=df_sandy, color="salmon")

# creating the legend, using pyplot
topbar = plt.Rectangle((0,0),1,1, fc="sage", edgecolor='none')
bottombar = plt.Rectangle((0,0),1,1, fc="salmon", edgecolor='none')
l = plt.legend([topbar, bottombar], ['Heat Wave Chicago', 'Sandy'], loc=1, ncol=2, prop={'size':10})
l.draw_frame(False)

# to rotate the x labels in order to fit in the graph
for item in ax.get_xticklabels():
    item.set_rotation(90)
```

In order to make a comparison between the two disasters, we have chosen a stacked bar chart to show the best result. For the Heat wave we have chosen the red color because there are much more victims than the Sandy(green). With the length of the bar we show the number of deaths, which will help the reader to interpret data.
